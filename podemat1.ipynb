```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
url ="https://docs.google.com/spreadsheets/d/1rtF7MFFn-YZ2edQT8fa0UKRyHtOcG5v1tmYxIn-oKD8/export?format=csv"

df = pd.read_csv(url)
df.columns = ['Dia', 'Previsao', 'Real']
```

```python
media_prev = df[ 'Previsao'].mean()
desvio_prev = df['Previsao'].std(ddof=0)

media_real = df['Real'].mean()
desvio_real = df['Real'].std(ddof=0)

print(f'Média da previsão: {media_prev:.2f}')
print(f'Desvio padrão da previsão: {desvio_prev:.2f}')
print(f'Média dos valores reais: {media_real:.2f}')
print(f'Desvio padrão dos valores reais: {desvio_real:.2f}')
```

```python
# Supondo que você já tenha carregado o DataFrame e calculado média e desvio padrão
media_real = df['Real'].mean()
desvio_real = df['Real'].std(ddof=0)

# Histograma com seaborn
sns.histplot(df['Real'], bins=10, stat='density', color='skyblue', edgecolor='black')

# Curva Gaussiana manual
x = np.linspace(df['Real'].min(), df['Real'].max(), 100)
gauss = (1 / (desvio_real * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - media_real) / desvio_real) ** 2)

plt.plot(x, gauss, 'r--', label='Distribuição Normal')

# Estilização
plt.title('Histograma das Vendas Reais com Curva Gaussiana')
plt.xlabel('Vendas Reais')
plt.ylabel('Densidade')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import norm
prob = norm.cdf(130, loc=media_prev, scale=desvio_prev)
print(f'A probabilidade de vender < 130 unidades é: {prob:.4f}')
```
